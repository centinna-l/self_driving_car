```python
import numpy as np
import matplotlib.pyplot as plt
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
from keras.datasets import mnist
from keras.utils.np_utils import to_categorical
import random
import tensorflow as tf
print("All imports done")
```

```python
np.random.seed(0)
```

```python
(x_train,y_train),(x_test,y_test)=mnist.load_data()
```

```python
assert(x_train.shape[0]==y_train.shape[0]),"The number of images is not equal to the number of labels"
assert(x_test.shape[0]==y_test.shape[0]),"The number of images is not equal to the number of labels"
assert(x_train.shape[1:]==(28,28)),"The dimensions of the image are not equal to the 28x28"
assert(x_test.shape[1:]==(28,28)),"The dimensions of the image are not equal to the 28x28"
```

```python
num_of_sample=[]
cols=5
no_of_classes=10

figs,axs=plt.subplots(nrows=no_of_classes,ncols=cols, figsize=(5,10))
figs.tight_layout()
for i in range(cols):
    for j in range(no_of_classes):
        x_selected=x_train[y_train==j]
        axs[j][i].imshow(x_selected[random.randint(0, len(x_selected - 1)), :, :], cmap=plt.get_cmap("gray"))
        axs[j][i].axis("off")
        if i==2:
            axs[j][i].set_title(str(j))
            num_of_sample.append(len(x_selected))
```

```python
print(num_of_sample)
plt.figure(figsize=(12,4))
plt.bar(range(0,no_of_classes), num_of_sample)
plt.title("Distribution of training Datasets")
plt.xlabel("Class number")
plt.ylabel("No of Images")
```

```python
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)
```

```python
x_train = x_train/255
x_test = x_test/255
```

```python
num_of_pixels = 784
x_train = x_train.reshape(x_train.shape[0], num_of_pixels)
x_test = x_test.reshape(x_test.shape[0], num_of_pixels)
print(x_train.shape)
print(x_test.shape)
```

```python
def create_model():
    model = Sequential()
    model.add(Dense(10, input_dim=num_of_pixels, activation="relu"))
    model.add(Dense(10, activation="relu"))
    model.add(Dense(10, activation="relu"))
    model.add(Dense(30, activation='relu'))
    model.add(Dense(no_of_classes, activation=tf.nn.softmax))
    model.compile(Adam(0.01), loss="categorical_crossentropy", metrics=['accuracy'])
    return model
```

```python
model = create_model()
print(model.summary())
```

```python
history = model.fit(x_train, y_train,verbose=1,validation_split=.1,batch_size=200, epochs=10, shuffle=1)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.xlabel('Epoch')
plt.legend(['acc','val_acc'])
plt.title('Loss')
```

```python
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.xlabel('Epoch')
plt.legend(['acc','val_acc'])
plt.title('Accuracy')
```

```python
score = model.evaluate(x_test, y_test, verbose=0)
print('Test Score: ', score[0])
print('Test Accuracy: ', score[1])
```
